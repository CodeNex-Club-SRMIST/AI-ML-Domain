Masking operations / MOST IMP

```python
mask = data['city']=='Hyderabad' # boolean series .
data[mask].shape # only 49 rows . (or) 49 matches have been played in Hyderabad .
# and if u want just the number of rows --->
data[mask].shape[0]
```

```python
def get_city_match_count(city):
  mask = data['city'] == city
  return data[mask].shape[0]
```

```python
get_city_match_count('Rajkot')
```

Using mutiple masks

```python
mask1 = data['city']=='Hyderabad'
mask2 = data['date']>'2010-01-01'

data[mask1 & mask2] # since both mask1 and mask2 are booleans
```

```python
data[mask1 & mask2].shape[0]
```

.value_counts() function is used to find out the frequency of all the categories in that particular column .

***Simply put it can compute how many times a particular element has occured in the column .***

```python
data['winner'].value_counts()  # here the elements became the index and the count is the output paired with it
```

```python
data['team1'].value_counts()+data['team2'].value_counts()
```

```python
(data['team1'].value_counts()+data['team2'].value_counts()).sort_values()
```

```python
(data['team1'].value_counts()+data['team2'].value_counts()).sort_values(ascending=False)
```
